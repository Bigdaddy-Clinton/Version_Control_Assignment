### Write a program that determines if a year entered by the user is a leap year. 
### (Hint: A leap year is divisible by 4, except for century years which must be divisible by 400).

```python
year = int(input("state a year: "))


if year % 400 == 0:
    print(f"This year, {year}, is a leap year")

elif year % 100 == 0:
    print(f"This year, {year}, is not a leap year")

elif year % 4 == 0:
    print(f"This year, {year}, is a leap year")

else:
    print(f"This year, {year}, is not a leap year")
```

### Create a simple calculator that takes two numbers and an operator (+, -, *, /) as input and performs the corresponding operation.

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter the operator (+, -, *, /): ")

if operator == "+":
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")
elif operator == "-":
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")
elif operator == "*":
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")
elif operator == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero.")
        exit(1)
    result = num1 / num2
    print(f"{num1} / {num2} = {result}")
else:
    print(f"Error: '{operator}' is not a supported operator. Use +, -, *, or /.")
```

### Write a program that categorizes a person by age: child (0-12), teenager (13-19), adult (20-64), or senior (65+).

```python
age = int(input("Enter the person's age (in years): "))

if age < 0:
    print("Error: Age cannot be negative.")
    exit(1)

if 0 <= age <= 12:
    category = "child"
elif 13 <= age <= 19:
    category = "teenager"
elif 20 <= age <= 64:
    category = "adult"
else:
    category = "senior"

print(f"At {age} years old, the person is categorized as a {category}.")
```

### Create a program that determines if a triangle is equilateral (all sides equal), isosceles (two sides equal), or scalene (no sides equal) based on three side lengths.

```python
side1 = float(input("Enter the length of side 1: "))
side2 = float(input("Enter the length of side 2: "))
side3 = float(input("Enter the length of side 3: "))

if side1 <= 0 or side2 <= 0 or side3 <= 0:
    print("Error: Side lengths must be positive.")
    exit(1)

if (side1 + side2 <= side3) or (side1 + side3 <= side2) or (side2 + side3 <= side1):
    print("Error: The given lengths do not form a valid triangle.")
    exit(1)

if side1 == side2 == side3:
    triangle_type = "equilateral"
elif side1 == side2 or side1 == side3 or side2 == side3:
    triangle_type = "isosceles"
else:
    triangle_type = "scalene"

print(
    f"A triangle with sides {side1}, {side2}, and {side3} is {triangle_type}."
)
```

### Write a program that checks if a string contains only digits, only letters, or a mix of both.

```python

s = input("Enter a string: ")

if s.isdigit():
    print("The string contains only digits.")
elif s.isalpha():
    print("The string contains only letters.")
else:
    has_digit = any(char.isdigit() for char in s)
    has_letter = any(char.isalpha() for char in s)
    
    if has_digit and has_letter:
        print("The string contains a mix of letters and digits.")
    else:
        print("The string contains characters other than letters and digits.")
```

### Write a program that checks if a name starts with a vowel or consonant.

```python
name = input("Enter a name: ").strip()
if len(name) == 0:
    print("Error: You must enter at least one character.")
    exit(1)

first_char = name[0].lower()

if not first_char.isalpha():
    print("Error: The first character is not a letter.")
    exit(1)

vowels = {"a", "e", "i", "o", "u"}

if first_char in vowels:
    print(f'"{name}" starts with a vowel ({name[0]}).')
else:
    print(f'"{name}" starts with a consonant ({name[0]}).')
```

### Write a program that removes all duplicate values from a list and prints the unique elements.

```python
elements_str = input("Enter elements separated by spaces: ").strip()

if len(elements_str) == 0:
    items = []
else:
    items = elements_str.split()

unique_items = []
for x in items:
    if x not in unique_items:
        unique_items.append(x)

print("Original list:", items)
print("Unique elements:", unique_items)
```

### Write a program that finds the most common element in a list and how many times it appears.

```python
elements_str = input("Enter list elements separated by spaces: ").strip()

if len(elements_str) == 0:
    print("Error: You must enter at least one element.")
    exit(1)
else:
    items = elements_str.split()

freq = {}
for x in items:
    freq[x] = freq.get(x, 0) + 1

most_common = None
max_count = 0

for element, count in freq.items():
    if count > max_count:
        most_common = element
        max_count = count
print(f"The most common element is '{most_common}' and it appears {max_count} time(s).")
```

### Create a program that counts the number of vowels and consonants in a sentence.

```python
sentence = input("Enter a sentence: ")
vowels = {"a", "e", "i", "o", "u"}

vowel_count = 0
consonant_count = 0

for char in sentence:
    if char.isalpha():
        # Convert to lowercase for a case-insensitive check
        lower_char = char.lower()
        if lower_char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print(f"Vowels: {vowel_count}")
print(f"Consonants: {consonant_count}")
```

### Write a program that finds the longest word in a sentence entered by the user.

```python
sentence = input("Enter a sentence: ").strip()

if len(sentence) == 0:
    print("Error: You must enter at least one character.")
    exit(1)

words = sentence.split()

longest_word = words[0]
for w in words[1:]:
    if len(w) > len(longest_word):
        longest_word = w

print(f'The longest word is "{longest_word}" with {len(longest_word)} character(s).')
```

### Create a program that sums user-entered numbers until the total exceeds 100, then stops and reports how many numbers were entered.

```python


total = 0.0
count = 0

while total <= 100:
  
    num_str = input("Enter a number: ")

    try:
        num = float(num_str)
    except ValueError:
        print("Error: Please enter a valid number.")
        exit(1)
  
    total += num
    count += 1


print(f"The total exceeded 100 after {count} number(s).")
```

### Create a program that prints the Fibonacci sequence up to a specific count provided by the user.

```python
count_str = input("Enter the number of Fibonacci terms to print: ").strip()

try:
    n = int(count_str)
    if n <= 0:
        print("Error: Please enter a positive integer.")
        exit(1)
except ValueError:
    print("Error: That is not a valid integer.")
    exit(1)

a, b = 0, 1

print("Fibonacci sequence:")
for _ in range(n):
    print(a)
    a, b = b, a + b
```

### Write a program that prints the multiplication table of a number entered by the user up to 10.

```python
num_str = input("Enter a number: ").strip()

try:
    num = int(num_str)
except ValueError:
    print("Error: Please enter a valid integer.")
    exit(1)

print(f"Multiplication table for {num}:")
for i in range(1, 11):
    product = num * i
    print(f"{num} x {i} = {product}")
```
